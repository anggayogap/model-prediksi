```python
#membaca/memasukkan library yang dibutuhkan
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#membaca file dataset

kelapasawit = pd.read_excel('datasetsawit.xlsx')
```

```python
#menampilkan data yang telah terbaca

kelapasawit
```

```python
#menghitung jumlah baris dan kolom pada data

kelapasawit.shape
```

```python
#menghitung nilai pada kesimpulan "1 & 0", 1 = pengepulan tepat waktu, 0 = pengepulan ditunda

kelapasawit['KESIMPULAN'].value_counts()
```

```python
#memisahkan data x = atribut, dan y = label (axis=1 adalah array 2 dimensi terdapat x & y)

x = kelapasawit.drop (columns='KESIMPULAN', axis=1)
y = kelapasawit['KESIMPULAN']
```

```python
#print untuk menampilkan data x & y

print(x)
```

```python
print(y)
```

```python
# untuk membuat rata-rata 0 dan variasi 1
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
#standarisasi data supaya range nilai sama
standarized_data = scaler.transform(x)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = kelapasawit['KESIMPULAN']
```

```python
print(x)
print(y)
```

```python
#spliting data (memecah data training = 80% dan testing 20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#untuk data yang dapat dipisahkan secara linear, 
#dataset dapat diklasifikasi menjadi dua kelas dengan menggunakan sebuah garis lurus tunggal

classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
#menguji akurasi prediksi data training

x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('akurasi data training adalah = ', training_data_accuracy)
```

```python
#menguji akurasi prediksi data testing

x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('akurasi data testing adalah = ', test_data_accuracy)
```
